# **NUMPY PROBLEMS**

```python
import numpy as np
import pandas as pd

df=pd.read_csv("csvfileforict.csv")
mpg_arr1=df["mpg"].values
print(mpg_arr1)
```

```python
#1)BASIC ARRAY OPERATIONS
mean_mpg=np.mean(mpg_arr1)
median_mpg=np.median(mpg_arr1)
sd_mpg=np.std(mpg_arr1)
count=(df["mpg"]>25).sum()

print("Mean of mpg:",mean_mpg)
print("Median of mpg:",median_mpg)
print("Standard deviation of mpg:",sd_mpg)
print("No:of cars with mpg greater than 25:",count)
```

```python
#2)FILTERING
cars_6=df[df["cylinders"]>6]["car name"]
print(cars_6)
cars_6list=df[df["cylinders"]>6]["car name"].tolist()
print(cars_6list)
```

```python
#3)BROADCASTING
df['horsepower'] = pd.to_numeric(df['horsepower'],errors='coerce')
hp_array=df["horsepower"].to_numpy()
mean_hp=np.nanmean(hp_array)
hp_array=np.where(np.isnan(hp_array),mean_hp,hp_array)
new_hp_array=hp_array*1.10
print(new_hp_array[:10])
```

```python
#4)BOOLEAN INDEXING
disp=df["displacement"].to_numpy()
origin=df["origin"].to_numpy()
avg_disp=np.mean(disp[origin==2])
print("Average Displacement:",avg_disp)
```

```python
#5)MEAN MPG BY CYLINDERS
mpg=df["mpg"].to_numpy()
cyl=df["cylinders"].to_numpy()

for i in np.unique(cyl):
  print("Cylinders:",i)
  print("Mean mpg:",np.mean(mpg[cyl==i]))
```

# **PANDAS PROBLEMS**

```python
#1)BASIC EXPLORATION

print(df.head(10))
print(df.shape)
print(df.describe())
```

```python
#2)FILTERING & INDEXING
result=df[(df["model year"]==75)&(df["weight"]<3000)]
print(result[["car name","weight","mpg"]])
```

```python
#3)HANDLING MISSING DATA
print(df.isnull().sum())
median_hp=df['horsepower'].median()
df['horsepower']=df['horsepower'].fillna(median_hp)
```

```python
#4)DATA TRANSFORMATION

df["power_to_weight_ratio"]=df["horsepower"]/df['weight']

print(df[['horsepower','weight','power_to_weight_ratio']])
```

```python
#5)GROUP BY

grp=df.groupby('origin')['mpg'].mean()
print(grp)
```

```python
#6)VISUAL PREPARATION

vp=df.groupby("model year")[['mpg','weight','horsepower']].mean()
print(vp)
```

```python
#7)EXPORTING DATA

highmpg=df[df['mpg']>30][['mpg','cylinders','horsepower','weight']]
highmpg.to_csv("high_mpg_cars",index=False)
```
